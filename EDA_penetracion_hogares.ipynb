```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_penetracion_hogares = pd.read_excel('Data/Internet.xlsx', sheet_name='Penetracion-hogares')
```

```python
df_penetracion_hogares
```

```python
df_penetracion_hogares.columns
```

```python
# Agrupar por 'Año' y 'Provincia' y sumar los valores de las columnas de acceso
df_penetracion_hogares = df_penetracion_hogares.groupby(['Año', 'Provincia']).sum().reset_index()

# Mostrar los resultados
df_penetracion_hogares
```

```python
# Agrupar por año y trimestre y calcular la media de 'Accesos por cada 100 hogares'
df_penetracion_resumen = df_penetracion_hogares.groupby(['Año', 'Trimestre'])['Accesos por cada 100 hogares'].mean().reset_index()

# Mostrar el resultado
print(df_penetracion_resumen)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear el gráfico de líneas
plt.figure(figsize=(10, 6))
sns.lineplot(x='Año', y='Accesos por cada 100 hogares', hue='Trimestre', data=df_penetracion_resumen, marker='o')

# Configuración del gráfico
plt.title('Evolución de Accesos por cada 100 Hogares por Año y Trimestre')
plt.xlabel('Año')
plt.ylabel('Accesos por cada 100 Hogares (%)')
plt.legend(title='Trimestre')
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
